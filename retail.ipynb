```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('/retail_sales_dataset.csv')
df.head()
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.drop_duplicates(inplace=True)
```

```python
display(df.head())
```

```python
df.groupby(df['Date'].dt.to_period('M'))['Total Amount'].sum().plot()
```

```python
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True)
```

```python
# 📊 Step 4: Descriptive Statistics
print("Summary Statistics:")
print(df.describe(include='all'))
print("Missing Values:")
print(df.isnull().sum())
```

```python

# Step 5: Data Cleaning
# Convert 'Order Date' to datetime if present
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
display(df.head())
```

```python
# Sales trend over time (monthly)
monthly_sales = df['Total Amount'].resample('M').sum()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Top 10 products by sales
top_products = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', figsize=(4,5), title='Top 10 Products by Sales')
plt.ylabel('Total Amount')
plt.show()
```

```python
customer_freq = df['Customer ID'].value_counts().head(10)
customer_freq.plot(kind='bar', figsize=(10,5), title='Top 10 Customers by Purchase Frequency')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()

```
